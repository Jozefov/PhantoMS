```python
def extract_label(exp_dir):
    match = re.search(r'cut_tree_\d+', exp_dir)
    if match:
        return match.group(0)
    return os.path.basename(exp_dir)

def same_depth_label(exp_dir):
    exp_folder = os.path.basename(os.path.dirname(exp_dir)).lower()
    
    if "dreams" in exp_folder:
        label = "dreams_embeddings"
    else:
        label = "raw_spectra"
    
    if "bonus" in exp_folder:
        label += "_formula"

    return label
```
